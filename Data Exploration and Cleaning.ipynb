```python
import pandas as pd
import numpy as np
```

```python
#loading the training data
train_data = pd.read_csv("train.csv")
# Exploring the data
train_data.head()
```

```python
# Data Types
train_data.info
```

```python
train_data.describe()
```

```python
# Check the number of rows before removing duplicates
print("Number of rows before removing duplicates:", len(train_data))

# Remove duplicates based on all columns
df = train_data.drop_duplicates()

# Check the number of rows after removing duplicates
print("Number of rows after removing duplicates:", len(df))

# Save the cleaned dataset to a new file
train_data.to_csv('titanic_cleaned.csv', index=False)
```

```python
# Check the number of missing values in each column
print("Missing values before handling:")
print(train_data.isnull().sum())
```

```python
# Impute missing values in numerical columns with mean
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Fare'].fillna(df['Fare'].mean(), inplace=True)

# Impute missing values in categorical columns with mode
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
# Check the number of missing values after handling
print("\nMissing values after handling:")
print(df.isnull().sum())

# Save the cleaned dataset to a new file
df.to_csv('titanic_cleaned.csv', index=False)
```

```python
# Identify outliers using Tukey's fences method

# Calculate the interquartile range (IQR)
Q1 = df['Age'].quantile(0.25)
Q3 = df['Age'].quantile(0.75)
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

```

```python
# Find the outliers in the 'Age' column
outliers = df[(df['Age'] < lower_bound) | (df['Age'] > upper_bound)]

# Print the outliers
print("\nOutliers:")
print(outliers)
```

```python
# Handle outliers by replacing them with appropriate values

# Replace with median
median_age = df['Age'].median()
df.loc[(df['Age'] < lower_bound) | (df['Age'] > upper_bound), 'Age'] = median_age
```

```python
# Save the cleaned dataset to a new file
df.to_csv('titanic_cleaned.csv', index=False)
```

```python
from sklearn.preprocessing import MinMaxScaler

# Select the numerical columns to normalize
numerical_cols = ['Age', 'Fare']

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Normalize the numerical columns
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])

# Save the normalized dataset to a new file
df.to_csv('titanic_normalized.csv', index=False)
```

```python
from sklearn.preprocessing import OneHotEncoder

# Select the categorical columns to encode
categorical_cols = ['Sex', 'Embarked']

# Perform one-hot encoding
encoder = OneHotEncoder(sparse=False)
encoded_cols = pd.DataFrame(encoder.fit_transform(df[categorical_cols]))

# Replace the original categorical columns with the encoded columns
df.drop(categorical_cols, axis=1, inplace=True)
df[encoded_cols.columns] = encoded_cols

# Save the encoded dataset to a new file
df.to_csv('titanic_encoded.csv', index=False)
```
